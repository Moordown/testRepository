```python
f = open("5", "r")
testCase5 = json.loads(f.read())
for alg in testCase5["algRes"]:
    axies = getTwoArrayByAlg(testCase5["algRes"][alg])
    plt.plot(axies[0], axies[1], 'b-', label=alg)
    plt.legend()
    plt.show()
```

```python
def sortedByVal(pair):
    return int(pair[0]);

def getTwoArrayByAlg(alg):
    xArg = []
    yArg = []
    lineArg = []
    for x in alg:
        lineArg.append((x, alg[x]))
    #print(lineArg);
    lineArg.sort(key=sortedByVal)
    #print(lineArg);
    for x in lineArg:
        xArg.append(x[0])
        yArg.append(x[1])
    return (xArg, yArg)
```
